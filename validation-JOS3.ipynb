```python
import jos3
import pandas as pd

model = jos3.JOS3(height=1.95, weight=88.6, age=25)
model.Ta = 27.8
model.RH = 40
model.PAR = 1.2
model.Va = 0.1
model.simulate(60)
model.Ta = 33.3
model.simulate(120)
model.Ta = 28.0
model.simulate(60)

df = pd.DataFrame(model.dict_results())  # Make pandas.DataFrame
```

```python
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['TskMean'], label='Mean Skin Temperature')
plt.plot(df.index, df['TcrPelvis'], label='Core Temperature (Pelvis)')

# Set the y-axis limits
plt.ylim(28, 40)

# Add labels and title
plt.xlabel('Time (minutes)')
plt.ylabel('Temperature (Â°C)')
plt.title('Skin and Core Temperature Over Time')
plt.legend()

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
